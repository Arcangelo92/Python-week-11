```python
#quali sono gli anni di inserimento presenti?
import pandas as pd
MCSV = pd.read_csv('dataset/Mappa-dei-pub-circoli-locali-in-Italia.csv', encoding="latin1", sep=';')
anni_inserimento = MCSV["Anno inserimento"].unique()
print (anni_inserimento)
```

```python
#quante attività ci sono nel quadrato di longitudine 9-10 e latitudine 45-46?
Long_lat = MCSV[(MCSV["Longitudine"] >= 9) & (MCSV["Longitudine"] <= 10) & (MCSV["Latitudine"] >= 45) & (MCSV["Latitudine"] <= 46)]
count = Long_lat.shape[0]
print (count)
```

```python
#quante attività ci sono nella provincia di Vicenza?
prov_Vi = MCSV["Provincia"] == "VICENZA"
count = prov_Vi.sum()
print ("attività nella provincia di Vicenza ->", count)
```

```python
#quante enoteche ci sono, e come si chiamano?
MCSV_cleaned = MCSV.dropna(subset=["Nome"])

enoteche = MCSV_cleaned["Nome"].str.contains("enoteca", case=False)
pub_enoteca = MCSV_cleaned[enoteche]["Nome"]
count = enoteche.sum()
print("Numero di pub con la parola 'enoteca' nel nome:", count)
print("Nomi dei pub che contengono la parola 'enoteca':")
print(pub_enoteca)
```

```python
#quali attività ci sono in Lazio e Abruzzo insieme?
LAZ_ABR = MCSV[(MCSV["Regione"] == "Abruzzo") & (MCSV["Regione"] == "Lazio")]
count = LAZ_ABR.shape[0]
print (count)
```

```python
def in_LAZ_ABR(group):
    return (group["Regione"] == "Lazio").any() and (group["Regione"] == "Abruzzo").any()

# Filtra il DataFrame in base alla funzione definita sopra
pub_LAZ_ABR = MCSV.groupby("Nome").filter(in_LAZ_ABR)

# Stampa i pub che si trovano sia nel Lazio che nell'Abruzzo
print("Pub che si trovano sia nel Lazio che nell'Abruzzo:")
print(pub_LAZ_ABR["Nome"].unique())
```

```python
## INSURANCE ##

import pandas as pd
pd.read_csv('dataset/insurance.csv')
ICSV = pd.read_csv('dataset/insurance.csv')

#Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna;
a = ICSV.shape
b = ICSV.columns
print (a,b)

#Quali sono le medie di charges rispetto a region? Ci sono differenze significative?
medie_spese_per_regione = ICSV.groupby('region')['charges'].mean().reset_index()
print("\nMedie delle spese (charges) rispetto alla regione:")
print(round(medie_spese_per_regione,2))

#E rispetto a smoker? E a sex?
medie_smsx_per_regione = ICSV.groupby(['sex', 'smoker']).size().unstack()
print(round(medie_smsx_per_regione,2))

#Quali sono i descrittori statistici di bmi? Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?
desc_bmi = ICSV['bmi'].describe()
print(desc_bmi)
quartili_bmi = ICSV['bmi'].quantile([0.25, 0.5, 0.75])
min_media_max_charges_per_quartile_bmi = ICSV.groupby(pd.cut(ICSV['bmi'], quartili_bmi, include_lowest=True)).agg({'charges': ['min', 'mean', 'max']})
print(min_media_max_charges_per_quartile_bmi)
```

```python
## IRIS ##

import pandas as pd
pd.read_csv('dataset/iris.csv')
IRCSV = pd.read_csv('dataset/iris.csv')
```

```python
IRCSV.columns
# La media della lunghezza dei petali di tutto il dataset 

MLP = IRCSV["petal_length"].mean()
print ("lunghezza media petali:" , round(MLP,2))

#• La media della lunghezza dei petali per ogni specie di Iris, utilizzando il metodo .groupby() 
MLP_per_specie = IRCSV.groupby('species')['petal_length'].mean()
print ("lunghezza media petali per specie:" , round(MLP_per_specie,2))

#• Media, minimo e massimo della larghezza dei sepali per ogni specie, utilizzando .groupby() e .agg()
MML = IRCSV.groupby('species')['sepal_width'].agg(['mean', 'min', 'max'])
print ("Media, minimo e massimo della larghezza dei sepali per ogni specie:" , round(MML,2))
```

```python
#ES
#Unire questi DataFrame in base alla colonna comune department_id, in modo da avere nel risultato informazioni sia sui dipendenti che sui dipartimenti, 
#usando la funzione .merge()
#Per ogni DataFrame, trasformare la colonna department_id nell'indice, facendo in modo che la modifica sia permanente;
#poi unire i due dataset mediante il metodo .join()
```

```python
#Per ogni DataFrame, trasformare la colonna department_id nell'indice, facendo in modo che la modifica sia permanente;
#poi unire i due dataset mediante il metodo .join()
import pandas as pd

employees_df = pd.DataFrame({ 
    'employee_id': [101, 102, 103, 104, 105], 
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'], 
    'department_id': [1, 2, 1, 2, 3] 
})

departments_df = pd.DataFrame({ 
    'department_id': [1, 2, 3], 
    'department_name': ['HR', 'IT', 'Finance'], 
    'location': ['New York', 'San Francisco', 'Chicago'] 
})

# Merge dei due DataFrame
merged_df = employees_df.merge(departments_df, on='department_id')

print(merged_df)

# Settare 'department_id' come indice per entrambi i DataFrame
employees_df.set_index('department_id', inplace=True)
departments_df.set_index('department_id', inplace=True)

# Unione utilizzando il metodo .join()
merged_df = employees_df.join(departments_df, lsuffix='_employee', rsuffix='_department')[["name", "department_name", "location"]]
print(merged_df)
```

```python
import os
import dotenv
from sqlalchemy import create_engine
import pandas as pd
dotenv.load_dotenv(override=True)
username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")
conn_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
db_engine = create_engine(conn_string)

query = "SELECT * FROM dimemployee"
df1 = pd.read_sql(query, db_engine)
db_engine.dispose ()
df1
```
