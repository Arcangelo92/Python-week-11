```python
# Dal database AdventureWorks estraiamo la tabella dimcustomer 
#• Trasformiamo i nomi dei clienti in modo che abbiano solo lettere minuscole, e i cognomi in modo che abbiano solo lettere maiuscole 
#• Sulla colonna EmailAddress, utilizzando il metodo .str.split(), estraiamo nome utente e dominio 
#• Sulla colonna Phone, estraiamo ogni parte del numero (ad es. da "1 (11) 500 555-0162" a ["1", "(11)", "500", "555-0162"]) 
#• Utilizzando il metodo .str.contains(), estraiamo tutti gli indirizzi e-mail che contengono il numero "21"
```

```python
import os
import dotenv
from sqlalchemy import create_engine
import pandas as pd
dotenv.load_dotenv(override=True)
username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")
conn_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
db_engine = create_engine(conn_string)

query = "SELECT * FROM dimcustomer"
df1 = pd.read_sql(query, db_engine)
db_engine.dispose ()

df1.columns 
```

```python
df1['FirstName'] = df1['FirstName'].str.lower()
df1['LastName'] = df1['LastName'].str.upper()
print(df1[['FirstName', 'LastName']])
```

```python
df1[['utente', 'Dominio']] = df1['EmailAddress'].str.split('@', expand=True)
#print(df1[['utente', 'Dominio']])
df1
```

```python
#• Sulla colonna Phone, estraiamo ogni parte del numero (ad es. da "1 (11) 500 555-0162" a ["1", "(11)", "500", "555-0162"]) 
pattern = r'(\d+) (\(\d+\)) (\d+) (\d+-\d+)'
phone_parts = df1['Phone'].str.extract(pattern)
phone_parts.columns = ['Paese', 'Area', 'Prefisso', 'Numero']
phone_parts
```

```python
#• Utilizzando il metodo .str.contains(), estraiamo tutti gli indirizzi e-mail che contengono il numero "21"
email_21 = df1[df1['EmailAddress'].str.contains("21")]
print(email_21['EmailAddress'])
```

```python
#Estraiamo tutti gli indirizzi e-mail che contengono il numero "20" oppure il numero "10"
email_20_10 = df1[df1['EmailAddress'].str.contains("20", "10")]
print(email_20_10['EmailAddress'])
```

```python
#Calcolare la lunghezza di ogni indirizzo e-mail ed estrarre i dieci più lunghi e i dieci più corti
df1['EmailLength'] = df1['EmailAddress'].str.len()
top_emails = df1.nlargest(10, 'EmailLength')
und_emails = df1.nsmallest(10, 'EmailLength')
print("Dieci indirizzi email più lunghi:")
print(top_emails[['EmailAddress', 'EmailLength']])
print("\nDieci indirizzi email più corti:")
print(und_emails[['EmailAddress', 'EmailLength']])
```

```python
#Modificare il dominio degli indirizzi e-mail da "adventure-works.com" a "aw-db.com" mediante il metodo .str.replace()
df1['EmailAddress'] = df1['EmailAddress'].str.replace('adventure-works.com', 'aw-db.com')
print(df1['EmailAddress'])
```

```python
#Dalla colonna AddressLine1 estraiamo tutti gli indirizzi che contengono la sottostringa "Street" 
Indirizzo = df1[df1['AddressLine1'].str.contains("Street")]
print(Indirizzo['AddressLine1'])
```

```python
import pandas as pd
FCSV = pd.read_csv('dataset/facebook.csv')
FCSV.columns
```

```python
#Con la funzione pd.to_datetime() convertiamo la colonna status_published in formato Timestamp
FCSV['status_published'] = pd.to_datetime(FCSV['status_published'])
print(FCSV['status_published'])
```

```python
FCSV['status_published'] = FCSV['status_published'].dt.strftime("%Y-%m-%d %H:%M:%S")
print(FCSV['status_published'])
```

```python
#Utilizzando gli attributi .dt.year , .dt.month , .dt.day , .dt.dayofweek , .dt.dayofyear, ottieniamo informazioni specifiche sulle date delle transazioni, come l'anno, il mese, il giorno della settimana, il giorno dell'anno, ecc.
FCSV['status_published'] = pd.to_datetime(FCSV['status_published'])

FCSV['anno'] = FCSV['status_published'].dt.year

FCSV['mese'] = FCSV['status_published'].dt.month

FCSV['giorno_settimana'] = FCSV['status_published'].dt.dayofweek

FCSV['giorno_anno'] = FCSV['status_published'].dt.dayofyear

print(FCSV[['status_published', 'anno', 'mese', 'giorno_settimana', 'giorno_anno']])
```

```python
#Estraiamo solo i post relativi al 2012

FCSV['anno'] = FCSV['status_published'].dt.year
status_2012 = FCSV[FCSV['anno'] == 2012]
status_2012
```

```python
#• Estraiamo solo i post relativi a maggio 2018
FCSV['mese'] = FCSV['status_published'].dt.month
FCSV['anno'] = FCSV['status_published'].dt.year
status_2018_05 = FCSV[(FCSV['anno'] == 2018) & (FCSV['mese'] == 5)]
status_2018_05
```

```python
#Confrontiamo il numero di post pubblicati nei weekend rispetto al numero di post pubblicati nel resto della settimana
FCSV['giorno_settimana'] = FCSV['status_published'].dt.dayofweek
post_weekend = FCSV[(FCSV['giorno_settimana'] == 5) | (FCSV['giorno_settimana'] == 6)]
post_settimana = FCSV[(FCSV['giorno_settimana'] != 5) & (FCSV['giorno_settimana'] != 6)]
print("Numero di post pubblicati nei weekend:", len(post_weekend))
print("Numero di post pubblicati nel resto della settimana:", len(post_settimana))
```

```python
#Troviamo il primo e ultimo post pubblicati in ogni anno
gruppo_anno = FCSV.groupby(FCSV['anno'])

def primo_e_ultimo(gruppo):
    primo_post = gruppo.iloc[0]['status_id']  
    ultimo_post = gruppo.iloc[-1]['status_id'] 
    return pd.Series([primo_post, ultimo_post], index=['Primo_post', 'Ultimo_post'])

primo_e_ultimo_post = gruppo_anno.apply(primo_e_ultimo)
primo_e_ultimo_post.reset_index(drop=True, inplace=True)

print(primo_e_ultimo_post)
```

```python
#Utilizzando il metodo .tz_localize(), aggiungiamo un fuso orario alle date delle transazioni
FCSV['status_published'] = FCSV['status_published'].dt.tz_localize('Europe/Rome', nonexistent='NaT')
FCSV
```

```python
#Calcoliamo la differenza di tempo tra due transazioni
FCSV = FCSV.sort_values(by='status_published')
FCSV['diff_tempo'] = FCSV['status_published'].diff()
FCSV1 = FCSV.dropna(subset=['diff_tempo'])
print(FCSV1[['status_published', 'diff_tempo']])
```

```python
#Quanti tipi di post ci sono? E quanti per ogni tipo?
numero_totale = len(FCSV['status_type'].unique())
conteggio_per_tipo = FCSV['status_type'].value_counts()
print("Numero totale di status_type di post:", numero_totale)
print("Conteggio per tipo di status_type:")
print(conteggio_per_tipo)
```

```python
## AUTOMOBILE ##

import pandas as pd
ACSV = pd.read_csv('dataset/automobile.csv')
ACSV.columns
```

```python
nulli = ACSV.isnull().any().any()

if nulli:
    nulli_per_colonna = ACSV.isnull().sum()
    print("Ci sono valori nulli nel dataset.")
    print("Numero di valori nulli per colonna:")
    print(nulli_per_colonna)
else:
    print("Non ci sono valori nulli nel dataset.")
```

```python
nullo = ACSV[ACSV['num-of-doors'].isnull()]
print(nullo.to_string())
```

```python
#Esaminando i dati, cerchiamo dei valori da sostituire ai valori nulli nella colonna num-of-doors
moda = ACSV['num-of-doors'].mode()[0]
ACSV['num-of-doors'] = ACSV['num-of-doors'].fillna(moda)
righe_valori_nulli = ACSV[ACSV['num-of-doors'].isnull()]
print(righe_valori_nulli)
```
